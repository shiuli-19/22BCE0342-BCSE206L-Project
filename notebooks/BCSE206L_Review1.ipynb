```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics as st
```

```python
# Load dataset
df = pd.read_csv('/content/GSS7402.csv')
```

```python
print("Missing values before filling:\n", df.isnull().sum())
```

```python
df['agefirstbirth'].fillna(df['agefirstbirth'].mean(), inplace=True)
```

```python
print("\n=== Column Summary ===")
for col in df.columns:
    data = df[col]
    print(f"column: {col}\t datatype: {type(data.iloc[0])}", end="")

    if np.issubdtype(data.dtype, np.number):
        print(f"\t Std. dev: {st.stdev(data)}\t mean: {st.mean(data)}", end="")
    else:
        unique_values = data.unique()
        if 'yes' in unique_values and 'no' in unique_values:
            yes_pct = 100 * len(data[data == 'yes']) / len(data)
            no_pct = 100 * len(data[data == 'no']) / len(data)
            print(f"\t Yes: {yes_pct:.2f}\t No: {no_pct:.2f}", end="")
    print()
```
